```python
Test = dt1.groupby('Survived').agg({'Age' : 'max'})
Test #Loggin the max age of the survivors column meaning that we have the maxiumun age of the those that survived and those that dont 0 and 1 
```

```python
AFm = dt1.groupby('Survived').agg( 
    {
        'Age' : ['median', 'mean'],
     'Fare' : ['median',]
    }
) 
AFm #Grouping and logging by the median and mean values of age and fare of the survival column
```

```python
pd.crosstab(dt1['Survived'], dt1['Sex']).plot.bar()
#Crosstabulating the columns of Survived with Sex resulting in a table that showcases the number of survivors by gender
```
